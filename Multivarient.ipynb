```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd
import io
mall_customer = pd.read_csv(io.BytesIO(uploaded['Mall_Customers.csv']))
print(mall_customer)
```

```python
#Converting Dataset into Dataframe using Pandas

mall_customer = mall_customer.drop('CustomerID',axis=1)
print(mall_customer)
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(mall_customer);
plt.show()
```

**Observation**: We can observe multiple variables plotted against each corresponding data points to compare and derive insights. We use Pair-plotting to acheive this.

```python
sns.heatmap(mall_customer.corr(),annot=True)
```

**Observation**: Graphical Representation of correlation values using colors to visualize the values in the matrix is observed. 

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(mall_customer, hue="Genre", size=3);
plt.show()
```

## Some more Data visualization techniques to get more insights from the db

**Box Plot** is simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. 

The lower and upper quartiles are shown as horizontal lines either side of the rectangle.

```python
sns.boxplot(x='Genre',y='Age', data=mall_customer)
plt.plot()
```

```python
sns.boxplot(x='Genre',y='Annual Income (k$)', data=mall_customer)
plt.plot()
```

```python
sns.boxplot(x='Genre',y='Spending Score (1-100)', data=mall_customer)
plt.show()
```

**Violin Plot** are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot.

```python
sns.violinplot(x="Genre", y="Annual Income (k$)", data=mall_customer, size=8)
plt.show()
```

```python
sns.violinplot(x="Genre", y="Spending Score (1-100)", data=mall_customer, size=8)
plt.show()
```
