# Bivariate Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
sns.set(style='whitegrid' ,palette='muted',color_codes=True)
```

```python
sns.scatterplot(x='Age',y='Annual Income (k$)',data=dataset)
plt.show
```

```python
sns.catplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=dataset)
plt.show()
```

```python
sns.pairplot(dataset[['Age', 'Spending Score (1-100)']])
plt.show()
```

**Observation**: The different plots such as catplot, scatterplot and pairplot against two variables "age" and "spending-score" has been plotted using seaborn data visualization.
